## Número de casos confirmados por municipio

```python
totales_confirmados=[]
num=[]
for i in range(0, 125):
    total_confirmado = len ( db_covid.loc[(db_covid['CVE_Mun_Res']== i+1)&(db_covid['RESULTADO_FINAL']== 'CONFIRMADO')])
    print ( municipios[i], ": ", total_confirmado)
    totales_confirmados.append (total_confirmado)
    num.append(i+1)
```

```python
plt.scatter(num, totales_confirmados)
plt.xlabel("Municipio")
plt.ylabel("Casos confirmados")
plt.show()
```
